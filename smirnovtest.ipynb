```python
import pandas as pd
import numpy as np
from scipy.stats import kstest
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('diamonds_new.xlsx')
```

```python
data
```

```python
statistic, pvalue = kstest(data['price'], 'norm' ) 
print(f'P-Value: {pvalue}')

if pvalue<0.05:
    print('The data doesnot follow specified distribution')
```

```python
import scipy as stats
from scipy.stats import kruskal
p_value = kruskal(data['price'], data['carat'], data['clarity'])

if p_value > 0.05:
    print('null hypothesis is true')
else:
    print('reject null hypothesis')
```

```python
# Group data by 'cut' and perform Kruskal-Wallis test
cuts = [data[data['cut'] == cut]['price'] for cut in data['cut'].unique()]

# Perform Kruskal-Wallis H Test
stat, p_val = stats.kruskal(*cuts)

# Check p-value
print(f"Kruskal-Wallis H Test p-value: {p_val}")
```
